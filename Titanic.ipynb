```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
from sklearn.model_selection import train_test_split

trainingset_df = pd.read_csv('train.csv')

trainingset_df = trainingset_df.dropna()

relevant_columns = ['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
trainingset_df = trainingset_df[relevant_columns + ['Survived']]
```

```python
trainingset_df.head(10)
```

```python
trainingset_df.groupby('Survived').count()

# The dataset is a bit inbalanced because we have a larger representation of the survivors
# Survived (contains your binary predictions: 1 for survived, 0 for deceased)
# However, we know that 712 people survived the sinking of the Titanic out of 2,207
```

```python
embarket_le = preprocessing.LabelEncoder()
embarket_le.fit(['C', 'Q', 'S'])

sex_le = preprocessing.LabelEncoder()
sex_le.fit(['female', 'male'])

trainingset_df['Embarked'] = embarket_le.transform(trainingset_df['Embarked']) 
trainingset_df['Sex'] = sex_le.transform(trainingset_df['Sex'])
```

```python
trainingset_df.head()
```

```python
from sklearn.model_selection import cross_val_score
X_train, X_test, y_train, y_test = train_test_split(trainingset_df[relevant_columns], trainingset_df['Survived'], test_size=0.25, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
model = gnb.fit(X_train, y_train)
scores = cross_val_score(model, X_test, y_test, cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.naive_bayes import ComplementNB 

gnb = ComplementNB()
model = gnb.fit(X_train, y_train)
scores = cross_val_score(model, X_test, y_test, cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
